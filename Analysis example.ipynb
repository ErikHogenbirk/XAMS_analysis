```python
# Notebookloader is needed to import ipython notebooks as python modules.
import NotebookLoader
import Function_definitions as fn

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import h5py
import units
# from numpy.lib.recfunctions import append_fields
```

```python
# Open the file
store = h5py.File('hdf5_example/150225_and_25bin_drift4.hdf5')
```

```python
# Slurp all events and peaks in to memory
events = store['Event'][:]
peaks = store['Peak'][:]
print("Number of events: ",len(events))
```

```python
# Group peaks by event
peaks_per_event = fn.group_by(peaks, 'Event')

# Get the main s1 / s2 of an event
# This funciton is in here because it needs the list defined above, and we may want to change the S1/S2 classification
def get_main_peak(event_nr, peak_type):
    pks = peaks_per_event[event_nr]
    pks = pks[pks['type'] == peak_type]
    if len(pks):
        return max(pks, key=lambda x:x['area'])
    else:
        return None
```

```python
# Make lists with relevant parameters
# Conventions here: - If we're talking about s1 or s2, start with s1_ or s2_
#                   - Then, add the parameter that it contains
#                   - If we have more than one element, use plural. 
#                        Example: s1_area is the area of ONE s1, s1_areas contains multiple
#                   - Use lower case

# Initialize all the lists to be filled. We use these lists later to generate awesome plots
s2_areas = []
s2_widths = []
s1_area_fractions = []
s2_area_fractions = []
drift_times = []

# Main loop: loop over all events and compute properties, but only if it contains at least one S1 and at least one S2
for e in np.arange(len(events)):
    main_s1 = get_main_peak(e, b's1')
    main_s2 = get_main_peak(e, b's2')
    if not (main_s1 and main_s2):
        # If event is missing one of the main peak types, move on
        continue
    # We have an S1 and an S2! Compute properties
    s2_widths.append(fn.dt*(main_s2['right']-main_s2['left']))
    s2_areas.append(main_s2['area'])
    drift_times.append(fn.dt * (main_s2['index_of_maximum'] - main_s1['index_of_maximum']))
    s1_area_fractions.append(main_s1['area_per_channel'][3] / main_s1['area'])
    s2_area_fractions.append(main_s2['area_per_channel'][3] / main_s2['area'])
    # This is how it should be computed, but doesn't work due to pax bug
#     s1_area_fractions.append(main_s1['area_fraction_top'])
#     s2_area_fractions.append(main_s2['area_fraction_top'])
drift_times = np.array(drift_times)
s2_widths   = np.array(s2_widths)
s2_areas    = np.array(s2_areas)
print('')
```

```python
# Drift time v. S2 width
# Good data: shows upward sloping trend
# In Xe gas: should be a function of square root plus some constant
plt.figure(figsize=(10,7))
plt.scatter(drift_times/ units.us, s2_widths / units.us,
            marker='.', edgecolor=None, alpha=0.4, s=50, c=np.log10(s2_areas));
plt.xlim(0,150)
plt.ylim(0,40)
plt.xlabel('Drift time (us)')
plt.ylabel('S2 width (MAD, us)')
plt.colorbar(label='log10 S2 area (pe)')
plt.show()
```

```python
# Get histogram with drift time v number of entries
# Should be rougly flat or have a bulge with collimated data
plt.figure(figsize=(10,7))
plt.hist(drift_times / units.us,bins=100);
plt.xlim(0,150)
plt.xlabel('Drift time (us)')
plt.ylabel('Number of entries');
```

```python
peak_area = []
peak_width = []
area_fraction = []
for event_iterator in peaks:
    peak_area.append(event_iterator['area'])
    peak_width.append(event_iterator['median_absolute_deviation'])
    area_fraction.append(event_iterator['area_fraction_top'])
peak_width = np.array(peak_width)
peak_area =  np.array(peak_area)
```

```python
#peaks.dtype.names
plt.figure(figsize=(10,7))
plt.scatter(drift_times/ units.us, s2_area_fractions,
            marker='.', edgecolor=None, alpha=0.4, s=50, c=np.log10(s2_areas));
plt.xlim(0,150)

plt.xlabel('Drift time (us)')
plt.ylabel('S2 top fraction area')
plt.colorbar(label='log10 S2 area (pe)')
plt.show()
```
