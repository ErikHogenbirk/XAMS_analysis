```python
# Notebookloader is needed to import ipython notebooks as python modules.
import NotebookLoader

import Function_definitions as fn
#import functions as fn

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import h5py
import units
# from numpy.lib.recfunctions import append_fields
```

```python
# Open the file
store = h5py.File('hdf5_example/150225_and_25bin_drift4.hdf5')
```

```python
# Slurp all events and peaks in to memory
events = store['Event'][:]
peaks = store['Peak'][:]
print("Number of events: ",len(events))
```

```python
# Group peaks by event
peaks_per_event = fn.group_by(peaks, 'Event')
# Get the main s1 / s2 of an event
# This funciton is in here because it needs the list defined above, and we may want to change the S1/S2 classification
def get_main_peak(event_nr, peak_type):
    pks = peaks_per_event[event_nr]
    pks = pks[pks['type'] == peak_type]
    if len(pks):
        return max(pks, key=lambda x:x['area'])
    else:
        return None
```

```python
# ==== Initialize lists ==== #
# Initialize all the lists to be filled. We use these lists later to generate awesome plots

# Make lists with relevant parameters
# These are lists for all ACCEPTED events, i.e. containint S1 and S2 and meeting some conditions
# Conventions here: - If we're talking about s1 or s2, start with s1_ or s2_
#                   - Then, add the parameter that it contains
#                   - If we have more than one element, use plural. 
#                        Example: s1_area is the area of ONE s1, s1_areas contains multiple
#                   - Use lower case

s1_areas = []
s2_areas = []
s1_widths = []
s2_widths = []
s1_area_fractions = []
s2_area_fractions = []
drift_times = []

# We count the numbers of events at different points, to see what we cut out
counter_events = len(events)
counter_s1_and_s2 = 0
counter_accepted = 0

# Main loop: loop over all events and compute properties, but only if it contains at least one S1 and at least one S2
for e in np.arange(len(events)):
    main_s1 = get_main_peak(e, b's1')
    main_s2 = get_main_peak(e, b's2')
    if not (main_s1 and main_s2):
        # If event is missing one of the main peak types, move on
        continue
    # We have an S1 and an S2! Increase counter and compute properties
    counter_s1_and_s2 +=1
    s1_area = main_s1['area']
    s1_width = fn.dt*(main_s1['right']-main_s1['left'])
    s2_width = fn.dt*(main_s2['right']-main_s2['left'])
    s2_area= main_s2['area']
    drift_time = fn.dt * (main_s2['index_of_maximum'] - main_s1['index_of_maximum'])
    area = peaks['area']
    s1_area_fraction = main_s1['area_per_channel'][3] / main_s1['area']
    s2_area_fraction = main_s2['area_per_channel'][3] / main_s2['area']
    # -- This is a good place for cuts --#
    #Cut for events happening at the cathode (blue events)
    if (s1_area_fraction<0.2 and s2_area_fraction <0.2) :
        continue    
    # If we pass the cut, store the parameters
    counter_accepted +=1
    s1_areas.append(s1_area)
    s1_widths.append(s1_width)
    s2_widths.append(s2_width)
    s2_areas.append(s2_area)
    drift_times.append(drift_time)
    areas.append(area)
    s1_area_fractions.append(s1_area_fraction)
    s2_area_fractions.append(s2_area_fraction)
    # This is how it should be computed, but doesn't work due to pax bug
#     s1_area_fractions.append(main_s1['area_fraction_top'])
#     s2_area_fractions.append(main_s2['area_fraction_top'])
drift_times = np.array(drift_times)
s2_widths   = np.array(s2_widths)
s2_areas    = np.array(s2_areas)

print('Number of events: %d (%.2f%%)' % (counter_events, counter_events/counter_events*100))
print('With S1 and S1:   %d (%.2f%%)' % (counter_s1_and_s2, counter_s1_and_s2/counter_events*100))
print('Accepted:         %d (%.2f%%)' % (counter_accepted, counter_accepted/counter_events*100))
```

```python
#==== Drifttime plots ==== #

# Good data: shows upward sloping trend
# In Xe gas: should be a function of square root plus some constant

plt.figure(figsize= (16,8))

# Simple drifttime plot
# Good statistics show flat distribution
plt.subplot(1,2,1)
plt.hist(drift_times/ units.us,  200)
plt.xlim(0,150)
plt.xlabel('Drift time (us)')
plt.ylabel('# counts')

plt.subplot(1,2,2)
#==== S2 broadening ==== #
plt.hist(s2_widths/ units.us,  200)
plt.xlim(0,100)
plt.xlabel('S2 width (MAD) (us)')
plt.ylabel('# counts')
plt.show()

# Drifttime vs width
# Good data: shows upward sloping trend
# In Xe gas: should be a function of square root plus some constant
plt.figure(figsize= (10,8))
plt.scatter(drift_times/ units.us, s2_widths / units.us,
            marker='.', edgecolor=None, alpha=0.4, s=50, c=np.log10(s2_areas));
plt.xlim(0,150)
plt.ylim(0,40)
plt.xlabel('Drift time (us)')
plt.ylabel('S2 width (right-left, us)')
zaxis = plt.colorbar()
zaxis.set_label('log10 S2 area (pe)', rotation=-270)

plt.show()

```

```python
# ===== S1 and S2 asymmetry plots ==== #

def AsymmetryPlot(AsymmetryType, nametitle):    
    plt.scatter(drift_times/ units.us, AsymmetryType,
                marker='.', edgecolor=None, alpha=0.4, s=50, c=np.log10(s2_areas));
    plt.xlim(0,140)
    plt.ylim(0,1)
    plt.xlabel('Drift time (us)')
    plt.ylabel('Fraction of light seen by top PMT')
    plt.title(nametitle)
    zaxis = plt.colorbar()
    zaxis.set_label('log10 S2 area (pe)', rotation=-270)    
    
# Amount of events with genuine S1 & S2
print(len(s1_area_fractions)/len(events)) 

plt.figure(figsize= (15,7))
    
plt.subplot(1,2,1)    
AsymmetryPlot(s1_area_fractions, 'S1 topfraction')
plt.subplot(1,2,2)
AsymmetryPlot(s2_area_fractions, 'S2 topfraction')
plt.show()

#blue events happen at the cathode. Linear course of the other events is good too
```
